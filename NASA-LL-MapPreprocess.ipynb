```python
import pandas as pd

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('all')
```

```python
# create preprocess_text function
def preprocess_text(text):

    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text
```

```python
df = pd.read_csv('Downloads/lesson_content.csv', encoding='utf8')
df.Paragraph = df.Paragraph.astype(str)

df
```

```python
df['Paragraph'] = df['Paragraph'].apply(preprocess_text)
df
```

```python
#make an ID column that identifies if it comes from the same website
url_unique = []
previous_url = ""
count_url = 0
for url in df.URL:
    if previous_url is url:
        url_unique.append(count_url)
    else:
        count_url += 1
        url_unique.append(count_url)
        previous_url = url

print(count_url)
print(len(url_unique))
print(url_unique)
```

```python
df.insert(1, "url_unique", url_unique)

df
```

```python
df.to_csv('./Downloads/processed.csv', index=False)
```
