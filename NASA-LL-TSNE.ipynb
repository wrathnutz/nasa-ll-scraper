```python

import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Downloads/processed.csv', encoding="utf8")
df.Paragraph = df.Paragraph.astype(str)
df
```

```python
vectorizer = CountVectorizer()
vectorizer.fit(df["Paragraph"].tolist())

print("Vocabulary: ", vectorizer.vocabulary_)
```

```python
vector = vectorizer.transform(df["Paragraph"].tolist())
print(vector.toarray())
```

```python
df_vector = pd.DataFrame(vector.toarray())
```

```python
scaled_data = StandardScaler().fit_transform(df_vector)
```

```python
tsne_2d = TSNE(n_components=2)
projections_2d = tsne_2d.fit_transform(scaled_data)
```

```python
fig = plt.figure()
ax = plt.axes()
sctt = ax.scatter(projections_2d[:, 0], projections_2d[:, 1], c=df["url_unique"], cmap='Spectral', s=5)
plt.gca().set_aspect('equal', 'datalim')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
plt.title('TSNE projection of the lessons learned', fontsize=24)
plt.show()
```

```python
tsne_3d = TSNE(n_components=3)
projections_3d = tsne_3d.fit_transform(scaled_data)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
sctt = ax.scatter3D(projections_3d[:, 0], projections_3d[:, 1], projections_3d[:, 2], c=df["url_unique"], cmap='Spectral', s=5)
plt.gca().set_aspect('equal', 'datalim')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
plt.title('TSNE projection of the lessons learned', fontsize=24)
plt.show()
```
